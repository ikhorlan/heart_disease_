```python
# Dropna method to drop data with missing variables
heart_dna=heart.dropna()
heart_dna.shape
```

```python
#Groupby method again to see final data by education group. 
heart_dna.groupby('education').agg({
    'age':'mean',
    'currentSmoker': 'mean',
    'cigsPerDay': 'mean',
    'BPMeds': 'mean',
    'prevalentHyp': 'mean',
    'diabetes': 'mean',
    'totChol': 'mean',
    'sysBP': 'mean',
    'BMI': 'mean',
    'heartRate': 'mean',
    'glucose': 'mean'  
})
```

```python
# Replacing qualitative variables into numbers
keyword_mappings = {
    'no': 0,
    'yes': 1
}

def i(row):
    for keyword, i in keyword_mappings.items():
        if keyword in row['prevalentStroke']:
            return i
    return None

heart_dna['prevalent_stroke'] = heart_dna.apply(i, axis=1)
```

```python
keyword_mappings = {
    'No': 0,
    'yes': 1
}

def n(row):
    for keyword, n in keyword_mappings.items():
        if keyword in row['Heart_ stroke']:
            return n
    return None

heart_dna['heart_stroke'] = heart_dna.apply(n, axis=1)
```

```python
#Replacing gender values with a dummy value
heartd=pd.DataFrame()
dcolumns=['Gender']
for column in dcolumns:
    df_dummy=pd.get_dummies(heart_dna[column], prefix=column, drop_first=True)
    heartd=pd.concat([heart_dna, df_dummy], axis=1)
    
print(heartd.shape)
```

```python
#Replacng education values with numbers
keyword_mappings = {'uneducated': 0, 'primaryschool': 1,'graduate': 2, 'postgraduate': 3}

heartd['edu'] = heartd.education.map(keyword_mappings)
```

<h4>XGBoost for heard disease<h4>

```python
xgb_model=xgb.XGBClassifier(objective='reg:squarederror',
    random_state=42)
```

```python
X = heartd.drop(columns=["Gender", "education", "prevalentStroke", 'Heart_ stroke', 'heart_stroke', "prevalent_stroke"])
y = heartd['heart_stroke']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, random_state=2)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
xgb_model.score(X_train, y_train)
```

```python
xgb_model.score(X_test, y_test)
```

```python
heart_feat=pd.DataFrame(xgb_model.feature_importances_, X_train.columns)
heart_feat.columns=['importance']
heart_feat=heart_feat.sort_values(by='importance', ascending=False).reset_index(drop=False)
heart_feat['cumulative_importance']=heart_feat['importance'].cumsum()
heart_feat
```
